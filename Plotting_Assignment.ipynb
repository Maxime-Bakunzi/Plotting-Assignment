```python
data_co2 = data['co2']
# print(data_co2.head(5))
sns.set_style('whitegrid')
sns.histplot(data_co2, bins = 25, color = 'grey')
py.xlabel('co2 Emission')
py.ylabel('Frequency')
py.title('Carbon Dioxide emitted by Vehicles')
py.savefig('co2_emission')
py.show()
```

**The Distribution of Carbon Dioxide(co2) by vehicles**
On the Y- axis shows the frequence of vehicles on which co2 was emitted while on X- axis is the emission of co2.

```python
data_ghg = data['ghgScore']
sns.set_style('ticks')
sns.histplot(data_ghg, bins = 28, color = 'darkgreen')
py.xlabel('GHG Score')
py.ylabel('Frequency')
py.title('Histogram of GHG Scores for Vehicles')
py.savefig('ghgScore')
py.show()
```

**Greenhouse Gas score**
On x- axis represents the GHG score of the vehicles while on Y- axis is the frequence of vehicles within the range of that GHG score.
# The vehicles of a ***score*** of **10** are the **cleanest**.

```python
data_fe = data['feScore']
sns.set_style('whitegrid')
sns.histplot(data_fe, bins = 30, color = 'darkblue')
py.xlabel('FE Score(L/100km)')
py.ylabel('Frequency')
py.title('Histogram of FE Scores for Vehicles')
py.savefig('feScore')
py.show()
```

**Fuel Economy(FE) score** on X- axis indicates the score of fuel of litres consumed within 100 km, the ***score*** 6 bellow is considered to be a **good** vehicle, on Y- axis represent the number of vehicles with a certain FE score.

```python
# separating numeric values from non numeric values
numeric_values = data.select_dtypes(include = ['number'])

corr_matrix = numeric_values.corr()
py.figure(figsize = (8, 8))
sns.heatmap(corr_matrix, cmap = "coolwarm", fmt = ' .2f')
py.xticks(rotation = 30)
py.title('Correlation Heat Map')
py.savefig('HeatMap')
py.show()
```

**The** **correlation** **heatmap** visually represents the pairwise relationships between numerical variables in the dataset. The intensity of colors in each cell indicates the strength and direction of correlation, with **lighter** shades representing positive correlations and **darker** shades representing negative correlations. Clusters of similarly colored cells highlight groups of strongly correlated variables, offering insights into patterns and dependencies within the data. This visualization aids in identifying potential relationships for further exploration and understanding the interplay between different variables in the dataset.
